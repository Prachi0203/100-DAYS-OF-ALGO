```python
class ProjectStructure2:
  
  def __init__(self,project_data):
    self.project_data = project_data
  
  def separatingProjects(self):
    projects = {}
    listOfProjects = []
    prev_head_conc = 0
    for i in self.project_data:  
      for j in i:
        #print(prev_head_conc)
        headings = project_head_obj.projHead_finder(j.strip())
        header = [i[2] for i in headings]
        #print(header)
        


        if header:
          if header[0].lower().strip()==j.lower().strip():
             continue

          head_conc = project_concept_obj.headings_concept(header)
          prev_head_conc = head_conc[0]

          #print('concept: ',head_conc[0])
          if head_conc[0] not in projects.keys():
            
            dictionary = {head_conc[0]:i}
            projects.update(dictionary)
            #projects[prev_head_conc].extend(i)
          else:
            listOfProjects.append(projects)
            projects = {}
        if not header:
          print('prev_had_conc: ',prev_head_conc)
          print('i: ',i)
        elif not header and prev_head_conc!=0:

          projects[prev_head_conc].extend(i)
    return listOfProjects
```
